```python
import pandas as pd
```

```python
import matplotlib as plt
```

```python
df = pd.read_csv("Electric_Production.csv")

print(df.head().to_markdown(index=True))
```

```python
df = df.rename(columns={
    "DATE": "Date",
    "IPG2211A2N": "Value"
})
```

```python
df.dtypes
```

```python
df["Date"] = pd.to_datetime(df["Date"])
print(df.head().to_markdown(index=True))
```

```python
df.dtypes
```

```python
df.set_index("Date", inplace= True)
print(df.head().to_markdown(index=True))
```

```python
df.plot()
```

```python
df.loc['2013'].plot()
```

```python
df.loc['2013'].plot(marker='o')
```

```python
df.loc['2004':'2014'].plot()
```

```python
df.resample('YE').mean()
```

```python
df.resample('YE').mean().plot(marker='o')
```

```python
df.rolling(window=6).mean().plot()
```
